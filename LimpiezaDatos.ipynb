# 2. Fichero pequeño pruebas

```python
import pandas as pd
```

```python
dataStream = pd.read_feather("data/anubisShorted.feather")
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## 2.1 Tren con muchos VINs que van a un mismo sitios -> Caso favorable

412062017-01-19

```python
# Este es un proceso normal y tipico
dl_A = dataStream.loc[(dataStream['ID'] == '412062017-01-19')] 
dl_A
```

```python
# Representacion de un VIN normal
dl2 = dataStream.loc[(dataStream['ID'] == '412062017-01-19') & (dataStream['LOGISTIC_OBJECT'] == 'VSSZZZ5FZHR016456') ] #2017-01-19 hasta 2019-01-06
dl2
```

492142020-02-20

```python
# Este es un proceso normal y tipico
dl_B = dataStream.loc[(dataStream['ID'] == '492142020-02-20')] #2017-01-19 hasta 2019-01-06
dl_B
```

  492762020-01-17

```python
# Este es un proceso normal y tipico
dl_C = dataStream.loc[(dataStream['ID'] == '492762020-01-17')] #2017-01-19 hasta 2019-01-06
dl_C
```

## 2.2 Tren con muchos VINs que van a sitios diferentes....

```python
# '412062017-01-29' presenta anomalias para cada VIN 
dl2 = dataStream.loc[(dataStream['ID'] == '412062017-01-29') ] #2017-01-19 hasta 2019-01-06
dl2 # en este hay problemas ya que 106489 106490 como ejemplo hacen eventos diferentes para un determinado VIN, pero no para el resto
```

Cada VIN tiene un destino diferente => tendrán individualmente diferentes eventos


```python
dl2 = dataStream.loc[(dataStream['ID'] == '412062017-01-29')  & (dataStream['LOGISTIC_OBJECT'] == 'VSSZZZ5FZHR021624')] #2017-01-19 hasta 2019-01-06
dl2
```

```python
dl2 = dataStream.loc[(dataStream['ID'] == '412062017-01-29')  & (dataStream['LOGISTIC_OBJECT'] == 'VSSZZZ5FZHR021698')] #2017-01-19 hasta 2019-01-06
dl2
```

```python
# Dentro de cad VIN, LLEGADA DE VAGONES Y PASO DE VAGONES ocurren en el mismo momento en localizaciones diferentes
# Lo que me dificultará para desglosar los VIN
dl2 = dataStream.loc[(dataStream['ID'] == '412062017-01-29')  & (dataStream['LOGISTIC_OBJECT'] == 'VSSZZZ5FZHR021218')] #2017-01-19 hasta 2019-01-06
dl2 
```

```python
dl2 = dataStream.loc[(dataStream['LOGISTIC_OBJECT'] == 'VSSZZZ5FZHR021218')] #2017-01-19 hasta 2019-01-06
dl2 
```

## 2.3 Guardo los ficheros de prueba

```python
df_short = dl_A.append(dl_B, ignore_index = True)
df_short = df_short.append(dl_C, ignore_index = True)
df_short.to_feather("data/anubisShorted_versionSimple.feather")
```
